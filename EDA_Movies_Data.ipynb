__Some additional information on Features/Columns__:

* **id:** The ID of the movie (clear/unique identifier).
* **title:** The Official Title of the movie.
* **tagline:** The tagline of the movie.
* **release_date:** Theatrical Release Date of the movie.
* **genres:** Genres associated with the movie.
* **belongs_to_collection:** Gives information on the movie series/franchise the particular film belongs to.
* **original_language:** The language in which the movie was originally shot in.
* **budget_musd:** The budget of the movie in million dollars.
* **revenue_musd:** The total revenue of the movie in million dollars.
* **production_companies:** Production companies involved with the making of the movie.
* **production_countries:** Countries where the movie was shot/produced in.
* **vote_count:** The number of votes by users, as counted by TMDB.
* **vote_average:** The average rating of the movie.
* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

## Find your next Movie

__Search 1: Science Fiction Action Movie with Bruce Willis (sorted from high to low Rating)__

```python
#Search 1
genres = movies.genres.str.contains('Action') & movies.genres.str.contains('Science Fiction')
actor = movies.cast.str.contains('Bruce Willis')
movies.loc[genres & actor, ['title','vote_average']].sort_values(by='vote_average',ascending=False)
```

__Search 2: Movies with Uma Thurman and directed by Quentin Tarantino (sorted from short to long runtime)__

```python
#Search 2
actor =  movies.cast.str.contains('Uma Thurman')
director = movies.director.str.contains('Quentin Tarantino')
movies.loc[director & actor, ['title','runtime']].sort_values(by='runtime',ascending=True)
```

__Search 3: Most Successful Pixar Studio Movies between 2010 and 2015 (sorted from high to low Revenue)__

```python
#Search 3
date = movies.release_date.between("2010-01-01", "2015-12-31")
studio = movies.production_companies.str.contains('Pixar').fillna(False)
movies.loc[studio&date,['title','revenue_musd']].sort_values(by='revenue_musd',ascending=False)
```

__Search 4: Action or Thriller Movie with original language English and minimum Rating of 7.5 (most recent movies first)__

```python
#Search 4
genres = movies.genres.str.contains('Action') | movies.genres.str.contains('Thriller')
language = movies['original_language']=='en'
rating = movies['vote_average']>=7.5
movies.loc[genres&language&rating&(movies.vote_count>50),['title','release_date']].sort_values(by='release_date',ascending=False)[0:20]
```

## Most Successful Actors (Part 1)

```python
movies.set_index("id",inplace=True)
```

```python
act = movies.cast.str.split('|',expand=True)
```

```python
act = act.stack().reset_index(level=1,drop=True).to_frame()
```

```python
act.columns = ['Actors']
```

```python
act = act.merge(movies[["title", "revenue_musd", "vote_average", "popularity"]],how='left',left_index=True,right_index=True)
```

## Most Successful Actors (Part 2)

```python
act.Actors.nunique()
#There are more than 200K actors
```

```python
#Actors featured in most movies
act.Actors.value_counts().head(20)
```

```python
plt.figure(figsize = (12, 8))
act.Actors.value_counts().head(20).plot(kind='bar', fontsize = 15)
plt.title("Most Active Actors",fontsize = 20)
plt.ylabel("Number of Movies", fontsize = 15)
plt.show()
```

```python
agg = act.groupby("Actors").agg(Total_Revenue = ("revenue_musd", "sum"), 
                               Mean_Revenue = ("revenue_musd", "mean"),
                               Mean_Rating = ("vote_average", "mean"), 
                               Mean_Pop = ("popularity", "mean"), 
                               Total_Movies = ("Actors", "count"))
```

```python
agg.nlargest(20,"Total_Movies")
```

```python
plt.figure(figsize = (12, 8))
agg.Total_Revenue.nlargest(10).plot(kind='bar', fontsize = 15)
plt.title("Total Revenue",fontsize = 20)
plt.ylabel("Revenue (in MUSD)", fontsize = 15)
plt.show()
```

```python
agg[agg.Total_Movies >= 10].nlargest(10, "Mean_Revenue")
```

```python
agg[agg.Total_Movies >= 10].nlargest(10, "Mean_Rating")
```

```python
agg[agg.Total_Movies >= 10].nlargest(10, "Mean_Pop")
```

## Most Successful Genres

```python
movies.release_date.dropna(inplace=True)
```

```python
movies['year'] = movies.release_date.apply(lambda x:x.year)
```

```python
movies.genres.dropna(inplace=True)
```

```python
genres = movies.genres.str.split('|',expand=True)
```

```python
genres = genres.stack().reset_index(level=1,drop=True).to_frame()
```

```python
genres.columns = ['genre']
```

```python
genres = genres.merge(movies[['title','budget_musd','revenue_musd','vote_count','vote_average','popularity','year','profit','ROI']],how='left',left_index=True,right_index=True)
```

```python
genres
```

```python
agg = genres.groupby('genre').agg({'budget_musd':['sum','mean'],'revenue_musd':['sum','mean'],'title':'count','vote_count':'count','vote_average':'mean','popularity':'mean'})
```

```python
agg.nlargest(20,('budget_musd','sum'))
#Genres with most allocated budgets
```

```python
agg.nlargest(20,('revenue_musd','sum'))
#Genres generating most revenue
```

```python
agg.nlargest(20,('title','count'))
#Genre-wise number of movies
```
